```python
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain.chat_models import init_chat_model
from langchain.tools import tool
load_dotenv(override=True)

model = init_chat_model(model='deepseek:deepseek-chat')
```

```python
from typing import Dict, Any

# Define tools
@tool
def query_customer_data(customer_id) -> Dict[str, Any]:
    """
    查询客户基本信息

    Args:
        customer_id: 客户ID，用于唯一标识客户
    Returns:
        包含客户基本的字典，如姓名，等级，加入日期等
    """
    return {"name": "张三", "level": "VIP", "join_date": "2023-01-15"}

@tool
def check_order_history(customer_id: str) -> Dict[str, Any]:
    """
    查询客户订单历史

    Args:
        customer_id: 客户ID，用于唯一标识用户

    Returns:
        包含用户订单历史的字典，如总订单数，总花费等
    """
    return {"total_order": 15, "total_spent": 25800.0}

@tool
def get_current_promotion() -> Dict[str, Any]:
    """
    获取当前可用的促销活动

    Returns:
         包含当前可用促销活动的字典，如活动名称、有效日期等
    """
    return {
        "promotion": ["老用户优惠", "会员专属折扣"],
        "valid_until": "2027-01-31"
    }
```

```python
from langchain.messages import HumanMessage
from rich import print as rprint

customer_service_agent = create_agent(
    model=model,
    tools=[query_customer_data, check_order_history, get_current_promotion],
)

messages = [
    HumanMessage("查询客户ID为cust1234的完整信息、历史订单和可用优惠")
]

for chunk in customer_service_agent.stream(
    {
        "messages": messages,
    },
    stream_mode="values"
):
    rprint(chunk)
    print("-"*30)
```

```python
for chunk in customer_service_agent.stream(
    {
        "messages": messages,
    },
    stream_mode="updates"
):
    rprint(chunk)
    print("-"*30)
```
