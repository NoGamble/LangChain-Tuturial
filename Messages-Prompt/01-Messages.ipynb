# Messages


```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
import os

from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

load_dotenv(override=True)

DEEPSEEK_API_KEY=os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL=os.getenv("DEEPSEEK_BASE_URL")

model = init_chat_model(
    model="deepseek:deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
)
```

```python
messages = [
    SystemMessage(content="You are a AI agent."),
    HumanMessage(content="1 + 1 = ?"),
    AIMessage(content="2"),
    HumanMessage(content="What question did I mention before?"),
]

model.invoke(messages)
```

```python
def keep_recent_message(messages, max_pairs = 3):
    """
    :param max_pairs: the maximum turns of messages to keep (one turn = user + assistant)
    """
    system_messages = [m for m in messages if m.get("role") == "system"]
    conversation_messages = [m for m in messages if m.get("role") != "system"]

    recent_messages = conversation_messages[-(max_pairs*2):]

    return system_messages + recent_messages
```
