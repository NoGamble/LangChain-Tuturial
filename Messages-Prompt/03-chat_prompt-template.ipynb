```python
# Messages

from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
import os
from rich import print as rprint
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate
)

load_dotenv(override=True)

DEEPSEEK_API_KEY = os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL = os.getenv("DEEPSEEK_BASE_URL")

model = init_chat_model(
    model="deepseek:deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
    temperature=1.5
)

chat_prompt_template = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template("You are a AI agent, and your name is {name}"),
    HumanMessage("How you doing"),
    AIMessage("I'm good."),
    HumanMessagePromptTemplate.from_template("{user_input}"),
])

prompt_value = chat_prompt_template.invoke(
    {
        "name": "Xiaozhi",
        "user_input": "2 + 2 =?"
    }
)
print(prompt_value)
rprint(model.invoke(prompt_value))
```

```python
template = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template("You are a {role}, your target customers are {audience}"),
    HumanMessagePromptTemplate.from_template("{task}"),
])

template = template.partial(role="guide", audience="visitors")

result1 = template.invoke(
    {
        "task": "introduce Forbidden City in Beijing"
    }
)

result2 = template.invoke(
    {
        "task": "introduce Chongqing"
    }
)
print(result1, result2)
```
