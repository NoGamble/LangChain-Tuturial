```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain.agents.middleware import SummarizationMiddleware
from langchain.agents import create_agent
from langchain.messages import HumanMessage, AIMessage, ToolMessage, SystemMessage

load_dotenv(override=True)

model = init_chat_model(
    model="deepseek:deepseek-chat",
    profile={"max_input_tokens": 128_000}
)

agent = create_agent(
    model="deepseek:deepseek-v4-flash",
    middleware=[
        SummarizationMiddleware(
            model=model,
            trigger=[
                ("tokens", 100),
                ("messages", 6),
                ("fraction", 0.001)
            ],
            keep=("messages", 2)
        ),
    ]
)

messages = [
    SystemMessage("You are a friendly AI agent assistant"),
    HumanMessage("Hello, this is NoGamble!"),
    AIMessage("Hello NoGamble, I'm a AI agent."),
    HumanMessage("Nice to meet you!"),
    AIMessage("You are in trouble now!"),
    HumanMessage("What do you mean?")
]

response = agent.invoke({
    "messages": messages
})

for msg in response["messages"]:
    msg.pretty_print()
```
