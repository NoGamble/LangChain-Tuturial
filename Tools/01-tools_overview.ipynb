```python
import os

from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_core.tools import tool

@tool
def get_weather(city: str) -> str:
    """
    :param city: name of the city to analyze
    :return: weather for the city
    """
    return city + "15 degree"
```

```python
get_weather.invoke({"city": "Beijing"})
```

```python
import os

load_dotenv(override=True)

DEEPSEEK_API_KEY = os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL = os.getenv("DEEPSEEK_BASE_URL")

model = init_chat_model(
    model = "deepseek-v4-flash",
    api_key = DEEPSEEK_API_KEY,
    base_url = DEEPSEEK_BASE_URL,
)

print(model.invoke("Hi!"))
```

```python
model_with_tools = model.bind_tools([get_weather])
response = model_with_tools.invoke("how is the weather in Beijing?")

if response.tool_calls:
    print("AI wants to use tools:", response.tool_calls)
else:
    print("AI answers directly:", response.content)
```
