```python
from anthropic import BaseModel
from langchain_core.utils.function_calling import convert_to_openai_tool
from rich import print as rprint

def get_weather(city: str) -> str:
    """
    inquire weather data by city

    Args:
        city (str): city name

    Returns:
        weather data
    """
    return f"{city} is sunny today!"

rprint(convert_to_openai_tool(get_weather))
```

# Use @Tool to define

```python
from langchain_core.tools import tool

@tool(name_or_callable="getWeather", description="Get weather data", parse_docstring=True)
def get_weather(city: str) -> str:
    """
    Get weather data by city

    Args:
        city (str): city name

    Returns:
        weather data
    """
    return f"{city} is sunny today!"

rprint(convert_to_openai_tool(get_weather))
```

# args_schema

```python
from langchain_core.tools import tool
from pydantic import BaseModel,Field
from typing import Literal

class WeatherInput(BaseModel):
    city: str = Field(description="city name", default="Beijing")
    unit: Literal["celsius", "fahrenheit"] = Field(default="celsius")
    include_forecast: bool = Field(
        description="whether include the weather forecast on next five days",
        default=False
    )

@tool(args_schema=WeatherInput)
def get_weather(city: str) -> str:
    """
    Get weather data by city
    """
    return f"{city} is sunny today!"

rprint(convert_to_openai_tool(get_weather))
```

# json schema

```python
json_schema = {
    'properties': {
        'city': {'default': 'Chongqing', 'description': 'city name', 'type': 'string'},
        'unit': {'default': 'celsius', 'enum': ['celsius', 'fahrenheit'], 'type': 'string'},
        'include_forecast': {
            'default': False,
            'description': 'whether include the weather forecast on next five days',
            'type': 'boolean'
        }
    },
    'type': 'object'
}

@tool(args_schema=json_schema)
def get_weather(city: str) -> str:
    """
    Get weather data by city
    """
    return f"{city} is sunny today!"

rprint(convert_to_openai_tool(get_weather))
```
