```python
from dotenv import load_dotenv
from langchain_deepseek import ChatDeepSeek
import os

from langchain_openai import ChatOpenAI
from langchain_openrouter import ChatOpenRouter
from pydantic import SecretStr

# Load the config
load_dotenv(override=True)

api_key_env = os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL = os.getenv("DEEPSEEK_BASE_URL")
DEEPSEEK_API_KEY = SecretStr(api_key_env) if api_key_env else None

llm_deepseek = ChatDeepSeek(
    model="deepseek-chat",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
)

message = llm_deepseek.invoke("introduce yourself in one sentence")
print(message)
```

```python
from langchain_community.chat_models import ChatZhipuAI

ZHIPUAI_API_KEY = os.getenv("ZHIPUAI_API_KEY")
ZHIPUAI_BASE_URL = os.getenv("ZHIPUAI_BASE_URL")

llm_zhipu = ChatZhipuAI(
    model="glm-5.2",
    api_key=ZHIPUAI_API_KEY,
    api_base=ZHIPUAI_BASE_URL,
)

message_2 = llm_zhipu.invoke("introduce yourself in one sentence")
print(message_2)
```

```python
# ChatOpenAI
from langchain_openai import ChatOpenAI
load_dotenv(override=True)
DEEPSEEK_API_KEY=os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL=os.getenv("DEEPSEEK_BASE_URL")

model = ChatOpenAI(
    model="deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
)

response = model.invoke("1 + 1 = ?")
print(response)
```

```python
# OpenRouter
from langchain_openrouter import ChatOpenRouter
import os
load_dotenv(override=True)
OPENROUTER_API_KEY=os.getenv("OPENROUTER_API_KEY")
OPENROUTER_API_BASE=os.getenv("OPENROUTER_API_BASE")

model = ChatOpenRouter(
    model="deepseek/deepseek-v4-flash",
    api_key=OPENROUTER_API_KEY,
    base_url=OPENROUTER_API_BASE,
)

response = model.invoke("1 + 1 = ?")
print(response)
```
