# invoke
## params of invoke
### text

```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
import os

load_dotenv(override=True)

DEEPSEEK_API_KEY=os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL=os.getenv("DEEPSEEK_BASE_URL")

model = init_chat_model(
    model="deepseek:deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
)
```

```python
response = model.invoke("Good morning!")
print(response)
```

### message

```python
messages = [
    {"role": "system", "content": "you are a professional math teacher"},
    {"role": "user","content": "Please explain to me what is Fibonacci series"}
]
response = model.invoke(messages)
print(response.content)
```

### multi-conversations

```python
messages = [
    {"role": "system", "content": "you are a professional math teacher"},
    {"role": "user","content": "1 + 1 = ?"},
    {"role": "assistant","content": "2"},
    {"role": "user","content": "what question did I mention just now"},
]

print(model.invoke(messages).content)
```

### Append Memory

```python
conversation = [
    {"role": "system", "content": "you are a AI agent assistant"},
    {"role": "user","content": "My name is NoGamble"},
]
response1 = model.invoke(conversation)
print(response1.content)

# Append agent's memories
conversation.append({"role": "assistant","content": response1.content})
conversation.append({"role": "user","content": "what is my name?"})

response2 = model.invoke(conversation)
print(response2.content)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

conversation = [
    SystemMessage(content="you are a AI agent assistant"),
    HumanMessage(content="My name is NoGamble"),
]
response1 = model.invoke(conversation)
print(response1.content)
```

```python
# Append agent's memories
conversation.append(AIMessage(content=response1.content))
conversation.append(HumanMessage(content="what is my name?"))

response2 = model.invoke(conversation)
print(response2.content)
```
